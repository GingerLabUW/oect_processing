### That should be displaying the transient data

The important columns are **Time (ms)**, **I_DS (A)**, and **I_DS_norm (a.u.)**. These correspond to
* Time : the times at which data are recorded
* I_DS (A): The drain-source current, in Amps
* I_DS_norm (a.u.): The drain-source current, normalized from 0 to 1

From here, we need to try two things:

1) Plot the data

2) Fit the data with a single exponential

Both are simple. The first is just ```transient.plot_current(df)``` where ```df``` is our data (stands for **DataFrame**)

The second is ```transient.fit_cycles(df, doping_time = 40, dedoping_time = 20)``` where the ```doping_time``` and ```dedoping_time``` are the time (in **seconds**) for the doping step and dedoping step per cycle. 40 here means 40 seconds, and 20 means 20 seconds. I.e. if our data are run for 1 minute per cycle, and in each cycle we dope for 40 seconds and then dedope for 20 seconds.

Let's try the first

````python
# Plot the data and show the plot_current docstring
transient.plot_current(df)
transient.plot_current?
# comment this second linout with a pound-sign if you don't want to see the docstring
````

## You can manually plot the data in df['I_DS (A)'] or df['I_DS_norm (a.u.)']

You should read the Pandas documentation for more detail, if you like. Then you could plot the specific columns.

Anyway, in the next cell, we fit our data

````python
doping_fits, dedoping_fits = transient.fit_cycles(df, doping_time = 40 , dedoping_time = 20, norm = False)
transient.fit_cycles?
# comment this second line out with a pound-sign if you don't want to see the docstring
````

## That's basically it!

You can now plot the time constant if you like

The Function ```fit_cycles``` returns the doping_fits and dedoping_fits. These contain the values found via curve-fitting for each cycle of data. 

Let's try plotting those

````python
fig, ax = plt.subplots(facecolor='white', figsize=(12, 8))
ax.plot([x[2] for x in doping_fits], 'rX', markersize=10, label='Doping taus')
ax2= ax.twinx()
ax2.plot([x[2] for x in dedoping_fits], 'bo', markersize=10, label='Dedoping taus')
ax.set_ylabel('Doping taus (s)')
ax2.set_ylabel('Dedoping taus (s)')
ax.set_xlabel('Cycle (#)')
ax.set_title('Time Constants per cycle')
````
